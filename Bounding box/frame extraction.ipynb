```python
# 프레임 정보 가져오기
import numpy
import cv2

def print_capture_properties(*args):
    capture = cv2.VideoCapture(*args)
    print('Created capture:', ' '.join(map(str, args)))
    print('Frame count:',
         int(capture.get(cv2.CAP_PROP_FRAME_COUNT)))
    print('Frame width:',
         int(capture.get(cv2.CAP_PROP_FRAME_WIDTH)))
    print('Frame height:',
         int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT)))
    print('Frame rate:', capture.get(cv2.CAP_PROP_FPS))
    

print_capture_properties('goodday.mp4')
```

```python
import cv2

capture = cv2.VideoCapture("sample.mp4")

while cv2.waitKey(33) < 0:
    if capture.get(cv2.CAP_PROP_POS_FRAMES) == capture.get(cv2.CAP_PROP_FRAME_COUNT): # 프레임 현재 구간과 전체 프레임이 같아졌다면
        capture.set(cv2.CAP_PROP_POS_FRAMES, 0)                                       # 프레임 현재 위치를 프레임 0으로 이동해라 

    ret, frame = capture.read()
    cv2.imshow("VideoFrame", frame)

capture.release()
cv2.destroyAllWindows()
```

```python
import json
with open('goodday.json','r',encoding='UTF-8')as f:
          json_data = json.load(f)
print(json.dumps(json_data, indent='\t'))  
```

```python
eye1_list = []
eyes1 = range(0, len(json_data[0]['data']))
for i in eyes1:
    eye1_a = json_data[0]['data'][i]['person'][0]['eyes1']
    line = []
    for j in range(0,4):
        eye1_b = eye1_a.split(',')[j]
        eye1_c = float(eye1_b)
        eye1_d = eye1_c / 3
        eye1_e = round(eye1_d)
        line.append(eye1_e)
    eye1_list.append(line)

print(eye1_list)
```

```python
eye2_list = []
eyes2 = range(0, len(json_data[0]['data']))
for i in eyes2:
    eye2_a = json_data[0]['data'][i]['person'][0]['eyes2']
    line = []
    for j in range(0,4):
        eye2_b = eye2_a.split(',')[j]
        eye2_c = float(eye2_b)
        eye2_d = eye2_c / 3
        eye2_e = round(eye2_d)
        line.append(eye2_e)
    eye2_list.append(line)

print(eye2_list)
```

```python
face_list = []
faces = range(0, len(json_data[0]['data']))
for i in faces:
    face_a = json_data[0]['data'][i]['person'][0]['face']
    line = []
    for j in range(0,4):
        face_b = face_a.split(',')[j]
        face_c = float(face_b)
        face_d = face_c / 3
        face_e = round(face_d)
        line.append(face_e)
    face_list.append(line)

print(face_list)
```

```python
mouth_list = []
mouths = range(0, len(json_data[0]['data']))
for i in mouths:
    mouth_a = json_data[0]['data'][i]['person'][0]['mouth']
    line = []
    for j in range(0,4):
        mouth_b = mouth_a.split(',')[j]
        mouth_c = float(mouth_b)
        mouth_d = mouth_c / 3
        mouth_e = round(mouth_d)
        line.append(mouth_e)
    mouth_list.append(line)

print(mouth_list)
```
